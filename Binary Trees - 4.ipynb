```python
header = "digraph sample {\n"
footer = "}\n"

prefix = "A"

def make_child(parent, child, prefix="A", left=True):
    return prefix + str(parent) + " -> " + prefix + str(child) + "\n" 

def make_node(ind, prefix="A", color="black", label=""):
    if label == "":
        label = str(ind)
    return prefix + str(ind) +"[label=" + label + " color=" + color + "]\n"

# make nodes
tree = ""
for i in range(10):
    tree += make_node(i)

tree += make_child(0, 1)
tree += make_child(0, 2)
tree += make_child(1, 3)
tree += make_child(1, 4)
tree += make_child(2, 5)
tree += make_child(3, 7)
tree += make_child(3, 8)
tree += make_child(5, 9)
tree += make_child(4, 6)

tree_source = header + tree + footer
gvz=graphviz.Source(tree_source)
# produce PDF
# gvz.view()
# print (gvz.source)
display(gvz)
```

```python
# tree walks
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None
        self.parent = None


def add_child(parent, val, left=True):
    child = Node(val)
    if left == True:
        parent.left = child
    else:
        parent.right = child
    child.parent = parent
    return child

###########################################
def print_inorder(tree):
    if tree == None:
        return
    if tree.left != None:
        print_inorder(tree.left)
    print(tree.val)
    if tree.right != None:
        print_inorder(tree.right) 

def print_nr_inorder(tree):
    if tree == None:
        return
    current = tree 
    prev = None
    while current != None:
        if prev == current.parent:
            prev = current
            current = current.left 
        if prev == current.left:
            print(current.val)
            prev = current
            current = current.right
        if prev == current.right:
            prev = current
            current = current.parent

def print_preorder(tree):
    if tree == None:
        return
    print(tree.val)    
    if tree.left != None:
        print_preorder(tree.left)
    if tree.right != None:
        print_preorder(tree.right) 

def print_postorder(tree):
    if tree == None:
        return
    if tree.left != None:
        print_postorder(tree.left)
    if tree.right != None:
        print_postorder(tree.right) 
    print(tree.val)    

tree = Node(0)
lnode = add_child(tree, 1)
rnode = add_child(tree, 2, left=False)

lnode = add_child(lnode, 5)
lnode = add_child(lnode, 6)
rnode = add_child(lnode, 12, left=False)
lnode = add_child(lnode, 8)
lnode = add_child(lnode, 10)

# 0 1 5 6 8 10 12 2
print("Preorder:")
print_preorder(tree)

# 10 8 6 12 5 1 0 2 
print("Inorder:")
print_inorder(tree)

# 10 8 12 6 5 1 2 0 
print("Postorder:")
print_postorder(tree)
```

LinkedList unsorted:
search, min/max, successor/predecessor  - O(n)
insert - O(1)

LinkedList sorted:
search - O(n/2) -> O(n)
min/max, successor/predecessor - O(1)
insert - O(n)

```python
from xmlrpc.client import MAXINT


class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.prev = None

def print_ds(shead):
    node = shead
    while node != None:
        print(node.val , " -> ", end="")
        node = node.next
    print("")

def insert(shead, val):
    new_node = Node(val)
    if shead != None:
        new_node.next = shead 
        shead.prev = new_node    
    return new_node 

# return the node the member is there or None otherwise
def search(shead, val):
    node = shead
    while node != None:
        if node.val == val:
            return node
        node = node.next
    return None

def successor(shead, found_node):    
    delta = -1 
    delta_node = None
    node = shead
    while node != None:
        # does the node has a chance to be a successor ? 
        if node.val > found_node.val:
            cur_delta = node.val - found_node.val
            if delta_node == None or cur_delta < delta:
                delta = cur_delta
                delta_node = node
        node = node.next
    return delta_node

def minimum(shead):
    cur_min = shead
    node = shead
    while node != None:
        if node.val < cur_min.val:
            cur_min = node
        node = node.next
    return cur_min


#######################
ds = None

ds = insert(ds, 4)
ds = insert(ds, 3)
ds = insert(ds, 8)
ds = insert(ds, 2)
ds = insert(ds, 100)

print_ds(ds)

found_node = search(ds, 2)
next_node = successor(ds, found_node)
# 3 
print(next_node.val)

found_node = search(ds, 100)
next_node = successor(ds, found_node)
# None
print(next_node)

min_node = minimum(ds)
print("Min:", min_node.val)
```
