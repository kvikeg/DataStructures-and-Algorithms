```python
import numpy as np
import pandas as pd
import math
import graphviz 
from IPython.display import display
```

```python
header = "digraph sample {\n"
footer = "}\n"

prefix = "A"

def make_child(parent, child, prefix="A", left=True):
    return prefix + str(parent) + " -> " + prefix + str(child) + "\n" 

def make_node(ind, prefix="A", color="black"):
    return prefix + str(ind) +"[label=" + str(ind) + " color=" + color + "]\n"

# make nodes
tree = ""
for i in range(10):
    tree += make_node(i)

tree += make_child(0, 1)
tree += make_child(0, 2)
tree += make_child(1, 3)
tree += make_child(1, 4)
tree += make_child(2, 5)
tree += make_child(3, 7)
tree += make_child(3, 8)
tree += make_child(5, 9)
tree += make_child(4, 6)

tree_source = header + tree + footer
gvz=graphviz.Source(tree_source)
# produce PDF
# gvz.view()
# print (gvz.source)
display(gvz)
```
