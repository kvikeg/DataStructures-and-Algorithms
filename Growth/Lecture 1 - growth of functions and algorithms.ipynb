Example:
$$ \frac{x^2}{2}  - 3 x = \Theta(x^2), c_1 = \frac{1}{14}, c_2 = \frac{1}{2}, x_0 = 7 $$ 

Proof:
$$ c_1 x^2 \leq  \frac{x^2}{2}  - 2 x \leq c_2 x^2, \forall x \geq x_0 $$ 
$$ c_1 \leq \frac{1}{2} - \frac{3}{x} \leq c_2 $$
$$ c_1 = \frac{1}{14}, c_2 = \frac{1}{2}, x_0 = 7 $$ 

```python
x = np.linspace(1, 50)
fx = (x**2) / 2 - 3 * x 
c1x = (x**2 / 14)
c2x = (x**2 / 2)

plt.figure(figsize=(16, 9))
sns.lineplot(x, fx, label="f")
sns.lineplot(x, c1x, label="lower bound")
sns.lineplot(x, c2x, label="upper bound")
```
