```python
%matplotlib inline
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
from IPython.display import IFrame
import scipy as sp

sns.set(style="darkgrid")
```

```python
# get 100 points from 0 to 100
x=np.linspace(1, 10, 10)

func1 = np.log2(x)
func2 = x
func3 = x* np.log2(x)
func4 = x**2
func5 = x**3
func6 = 2**x
func7 = 3**x
func8 = sp.special.factorial(x)
func9 = x**x

sns.plot(x, func1, label="func1")
plt.plot(x, func2, label="func2")
plt.plot(x, func3, label="func3")
plt.plot(x, func4, label="func4")
plt.plot(x, func5, label="func5")
plt.plot(x, func6, label="func6")
plt.plot(x, func7, label="func7")
plt.plot(x, func8, label="func8")
plt.plot(x, func9, label="func9")

plt.legend(loc='upper left')
plt.show()
```
