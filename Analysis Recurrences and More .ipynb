```python
# to sort an array "a" call it like: merge_sort(a, 0, len(a))
def merge_sort(a, left, right, print_prefix=""):
    print(print_prefix, "Called merge_sort(a, ", left, ",", right, ")" )
    if left < right:
        mid = math.floor((left + right) / 2.0)
        print(print_prefix, "mid = ", mid)
        temp_a = merge_sort(a, left, mid, print_prefix + "\t")
        print(print_prefix, "after first merge sort", temp_a)
        temp_a = merge_sort(temp_a, mid+1, right, print_prefix + "\t")
        print(print_prefix, "after second merge sort:", temp_a)
        return merge(temp_a, left, mid, right, print_prefix)
    else:
        return a
        
def merge(a, left, mid, right, print_prefix=""):
    print(print_prefix, "Called merge: a=", a, "left=", left, "mid=", mid, "right=", right)
    tempa = a[:]
    left_ind = left
    right_ind = mid+1
    temp_ind = left
    while left_ind <= mid or right_ind <= right:
        if left_ind > mid:
            tempa[temp_ind] = a[right_ind]
            right_ind +=1 
        elif right_ind > right or a[left_ind] < a[right_ind]:
            tempa[temp_ind] = a[left_ind]
            left_ind += 1
        else:
            tempa[temp_ind] = a[right_ind]
            right_ind +=1 
            
        temp_ind += 1

    print(print_prefix, "Merged: ", tempa)
    return tempa
```

## Example 1 (substitution):
$$
T(n) = \left\{
    \begin{array}\\
        T(1) & \mbox{if } \ n = 1 \\
        2 T(n/2) + n & \mbox{if } \ n > 1
    \end{array}
\right.
$$
## Guess:
$$
T(n) = n \log(n) + n.
$$
## Base: 
$$
n=1, n \log (n) + n = 1 = T(1).
$$
## Step:
Assume: $ T(m) = m \log(m) +m, \forall m < n$

$$ 
\begin{array}\\
T(n) =& 2 T(\frac{n}{2}) + n \\ 
& = 2 (\frac{n}{2} \log(\frac{n}{2}) + \frac{n}{2}) + n \\
& = n \log(\frac{n}{2}) + 2n \\
& = n \log(n) - n \log(2) + 2n \\
& = n \log (n) + n.
\end{array}
$$

## Example 2 (substitution): Guessing bounds
$$
T(n) = \left\{
    \begin{array}\\
        T(1) & \mbox{if } \ n = 1 \\
        2 T(n/2) + n & \mbox{if } \ n > 1
    \end{array}
\right.
$$
## Guess: upper bound
$$
T(n) \leq cn \log(n), \forall n \geq 2.
$$
## Base: 
$$
n=2, T(2) = 2 T(1) + 2 = 4 \leq 2c \Rightarrow c \geq 2.
$$
## Step:
Assume: $ T(m) \leq c m \log(m), \forall 2 \leq m < n$

$$ 
\begin{array}\\
T(n) =& 2 T(\frac{n}{2}) + n \\ 
& = 2 (c\frac{n}{2} \log(\frac{n}{2}) + n \\
& = cn \log(\frac{n}{2}) + n \\
& = cn \log(n) - cn + n \\
& \leq cn \log (n) + n  \\
& if -cn + n \leq 0 \Rightarrow c \geq 1.\\
T(n) =& O(n\log(n))
\end{array}
$$
## Guess: lower bound
$$
T(n) \geq dn \log(n), \forall n.
$$
## Base: 
$$
n=1, T(1) = 1 \geq d log (1) = 0.
$$
## Step:
Assume: $ T(m) \geq d m \log(m), \forall 2 \leq m < n$

$$ 
\begin{array}\\
T(n) =& 2 T(\frac{n}{2}) + n \\ 
& \geq  2 (d\frac{n}{2} \log(\frac{n}{2}) + n \\
& = dn \log(\frac{n}{2}) + n \\
& = dn \log(n) - dn + n \\
& \geq dn \log (n) \\
& if -dn + n \geq 0 \Rightarrow d \leq 1.\\
T(n) =& \Omega(n\log(n)) \\
T(n) =& \Theta(n\log(n))
\end{array}
$$
